```python
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
```

```python
state_vector = qi.Statevector.from_label('00')
state_vector.draw('latex')
```

```python
qc = QuantumCircuit(2)
qc.h(0)
state_vector = qi.Statevector(qc)
state_vector.draw('latex')
```
